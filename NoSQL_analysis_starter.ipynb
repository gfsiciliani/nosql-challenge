### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}
results = db.establishments.find(query)

# Use count_documents to display the number of documents in the result
count = db.establishments.count_documents(query)
print("Number of establishments with a hygiene score of 20:", count)

# Display the first document in the results using pprint
first_document = db.establishments.find_one(query)
print(f'''
      ------------------
      - FIRST DOCUMENT -
      ------------------
      ''')
pprint(first_document)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'RatingValue': {'$gte': 4}}
ratingval_results = db.establishments.find(query)

# Use count_documents to display the number of documents in the result
count = db.establishments.count_documents(query)
print("Number of establishments with a RatingValue of 4 or greater:", count)

# Display the first document in the results using pprint
first_document = db.establishments.find_one(query)
print(f'''
      ------------------
      - FIRST DOCUMENT -
      ------------------
      ''')
pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame
ratingval_df = pd.DataFrame(ratingval_results)

# Display the number of rows in the DataFrame
print(f'Number of rows in df: {len(ratingval_df)}')

# Display the first 10 rows of the DataFrame
ratingval_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142
longitude = 0.08384000

# Query to match documents within a geographical range and with a RatingValue of 5
query = {
    '$match': {
        'geocode.longitude': {'$gte': longitude - degree_search, '$lte': longitude + degree_search},
        'geocode.latitude': {'$gte': latitude - degree_search, '$lte': latitude + degree_search},
        'RatingValue': 5
    }
}

# Sorting to order by hygiene score
sort_stage = {'$sort': {'scores.Hygiene': 1}}

# Limit stage to get only the top 5 results
limit = {'$limit': 5}

# Aggregation pipeline
pipeline = [query, sort_stage, limit]

# Execute the pipeline
results = list(establishments.aggregate(pipeline))

# Display the results
for establishment in results:
    print(establishment)
```

```python
# Convert result to Pandas DataFrame
df = pd.DataFrame(results)
df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Matching establishments with a hygiene score of 0
match_stage = {
    '$match': {'scores.Hygiene': 0}
}

# Grouping the matches by Local Authority and counting each group
group_stage = {
    '$group': {
        '_id': '$LocalAuthorityName',
        'count': {'$sum': 1}
    }
}

# Sorting the matches from highest to lowest count
sort_stage = {
    '$sort': {'count': -1}
}

# Create the pipeline
pipeline = [match_stage, group_stage, sort_stage]

# Execute the pipeline
results = list(db.establishments.aggregate(pipeline))

# Print number of documents in result
print("Number of Local Authorities with hygiene score of 0:", len(results))

# Print the first 10 results
for result in results[:10]:
    print(result)
```

```python
# Convert the result to a Pandas DataFrame
df2 = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(f'Number of rows in DataFrame: {len(df2)}')

# Display the first 10 rows of the DataFrame
df2.head(10)
```
